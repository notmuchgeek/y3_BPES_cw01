```python
def load_d(name):
    xy_data=np.loadtxt(PATH+str(name),skiprows=3,delimiter=',',dtype='str')

    x1=[]
    y1=[]
    for i in xy_data:
        x1.append(i[0])
        y1.append(float(i[1]))
    return np.array(x1), np.array(y1)
```

```python
date_divPay,div_divPay = load_d('div_payout.csv')



fig, ax = plt.subplots()
plt.title('DivPay')


ax.plot(div_divPay, linewidth=1,label = 'DivPay', marker='.')

ax.set_xlabel('Date')
ax.set_ylabel('divPay')

ax.grid(True, which='both')
ax.axhline(y=0, color='k')
ax.axvline(x=0, color='k')

#ax.set_xlim(0,0.3)

ax.legend()
```
