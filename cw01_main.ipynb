```python
%pylab inline
import pandas as pd
from scipy.signal import find_peaks
import plotly.graph_objects as go

import matplotlib.pyplot as plt
import numpy as np
```

```python
import os

#Universal path for the working dictionary
PATH = 'C:/Users/AlbertCielstian/Documents/GitHub/y3_BPES_cw01/data/'

files = os.listdir(PATH)

for file in files:
    print(file)
```

```python
def load_d(name):
    xy_data=np.loadtxt(PATH+str(name),skiprows=0)

    x1=[]
    y1=[]
    for i in xy_data:
        x1.append(i[0])
        y1.append(float(i[1]))
    return np.array(x1), np.array(y1)
```

## All data sources and variables

```python
g_year,g = load_d('div_growth_by_year.txt')
rm_year,rm = load_d('div_anual_total_return.txt')
div_year,div = load_d('div_by_year.txt')


# https://ycharts.com/indicators/reports/sp_500_returns
# https://www.multpl.com/s-p-500-dividend/table/by-year
# https://www.multpl.com/s-p-500-dividend-growth
```

## Div Growth (g)

```python
fig, ax = plt.subplots()

plt.title('Dividend Growth By Year')
ax.plot(g_year,g, linewidth=1,label = 'Div growth (g)', marker='.')



ax.set_xlabel('year')
ax.set_ylabel('div growth (g)')
```

## Market return (rm)

```python
fig, ax = plt.subplots()

plt.title('Dividend Anual Total Return')
ax.plot(rm_year,rm, linewidth=1,label = 'Markert Return (r_m)', marker='.')

ax.set_xlabel('year')
ax.set_ylabel('markert return ($r_m$)')
```

### Market return and dividend growth

```python
plt.rcParams['figure.figsize'] = [8, 6]

fig, ax1 = plt.subplots()

ax2 = ax1.twinx()
ax2.plot(g_year[:-3],g[:-3],color='g',marker='.',linestyle='-')
ax1.plot(rm_year,rm,color='r',marker='.',linestyle='--')

ax1.set_xlabel('Year')
ax2.set_ylabel('Div growth (g)', color='r')
ax1.set_ylabel('markert return ($r_m$)', color='g')

ax1.axhline(y=0, color='k')


plt.show()
```

## Div by year (Div)

```python
fig, ax = plt.subplots()

plt.title('div_by_year.txt')
ax.plot(div_year,div, linewidth=1,label = 'Dividend (Div)', marker='.')
ax.plot(div_year[2:8],div[2:8], linewidth=1,label = 'Dividend (Div)', 
        marker='*',color='r',linestyle='')

ax.set_xlabel('year')
ax.set_ylabel('Dividend (Div)')
```

### S&P 500 index and dividend by year

```python
index_year, index = load_d('s_p_500_historical_price.txt')


plt.rcParams['figure.figsize'] = [8, 6]

fig, ax1 = plt.subplots()

ax2 = ax1.twinx()
ax1.plot(index_year[:65], index[:65],color='g',marker='.',linestyle='-')
ax2.plot(div_year[:65],div[:65],color='r',marker='.',linestyle='--')

ax1.set_xlabel('Year')
ax1.set_ylabel('S&P 500 Index', color='g')
ax2.set_ylabel('Dividend by year (Div)', color='r')


plt.show()
```
